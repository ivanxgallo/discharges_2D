```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Cargar datos desde el archivo CSV
archivo_csv = 'density_1_400.csv'
matriz = np.loadtxt(archivo_csv, delimiter='\t')

plt.figure(dpi=200)
# Crear el mapa de calor
plt.imshow(matriz, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Valor')

# Mostrar el gr√°fico
plt.show()
```

```python
files = [f"density_1_{i*100}.csv" for i in range(1, 21)]
```

```python
files
```

```python
def load_matrix(csv_file):
    return np.loadtxt(csv_file, delimiter='\t')


matrices = [load_matrix(csv_file) for csv_file in files]
```

```python
from matplotlib.colors import LogNorm

# Create figure and axis
fig, ax = plt.subplots()
eps = 1e-3
vmin = 1e-3
vmax = 26
# Update function for the animation
def update(frame):
    ax.clear()
    ax.imshow(matrices[frame] + eps, cmap='viridis', interpolation='nearest', norm=LogNorm(vmin=vmin, vmax=vmax))
    ax.set_title(f'Frame {frame + 1}/{len(matrices)}')

# Create the animation
animation = FuncAnimation(fig, update, frames=len(matrices), interval=1000, repeat=False)

# Save the animation as a GIF file
animation.save('temporal_evolution.gif', writer='imagemagick', dpi=200)

# Show the animation (optional)
plt.show()
```

```python
import os
import re

def ordenar_archivos_por_density_nombre(archivo):
    # Utilizar expresi√≥n regular para extraer los valores de n y m
    match = re.match(r'density_(\d+)_(\d+)', archivo)
    
    if match:
        n = int(match.group(1))
        m = int(match.group(2))
        return n, m
    else:
        # Si no hay coincidencia, devolver una tupla vac√≠a
        return ()

# Obtener el directorio actual
directorio_actual = os.getcwd()

# Obtener la lista de archivos en el directorio actual
archivos = os.listdir(directorio_actual)

# Filtrar los archivos que comienzan con "density"
archivos_density = [archivo for archivo in archivos if archivo.startswith("density")]

# Ordenar la lista utilizando la funci√≥n personalizada
archivos_density_ordenados = sorted(archivos_density, key=ordenar_archivos_por_density_nombre)
```

```python
archivos_density_ordenados
```

```python
def load_matrix(csv_file):
    return np.loadtxt(csv_file, delimiter='\t')


matrices = [load_matrix(csv_file) for csv_file in archivos_density_ordenados]
```

```python
from matplotlib.colors import LogNorm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Create figure and axis
fig, ax = plt.subplots()
eps = 1e-3
vmin = 1e-3
vmax = 500
# Update function for the animation
def update(frame):
    ax.clear()
    ax.imshow(matrices[frame] + eps, cmap='viridis', interpolation='nearest', norm=LogNorm(vmin=vmin, vmax=vmax))
    ax.set_title(f'Frame {frame + 1}/{len(matrices)}')

# Create the animation
animation = FuncAnimation(fig, update, frames=len(matrices), interval=1000, repeat=False)

# Save the animation as a GIF file
animation.save('temporal_evolution4.gif', writer='imagemagick', dpi=200)

# Show the animation (optional)
plt.show()
```
